```python
import pandas as pd
iris = pd.read_csv('/content/drive/MyDrive/iris.csv')
iris.head()
```

```python
# Shape of the dataset
print("Shape of dataset:", iris.shape)

# Column names
print("Column names:", iris.columns.tolist())

# First few rows
iris.head()
```

```python
print("Dataset Info:")
iris.info()

print("\nSummary Statistics:")
iris.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Scatter plot between sepal_length and sepal_width
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=iris,
    x="SepalLengthCm",
    y="SepalWidthCm",
    hue="Species"
)
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

```python
iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].hist(
    figsize=(10, 8),
    bins=20,
    color='skyblue'
)
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
plt.title("Box Plots of Iris Dataset Features")
plt.show()
```
